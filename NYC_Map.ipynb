```python
import folium
import geopandas
import geopy
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
locator = Nominatim(user_agent= 'jchen2712@gmail.com')
```

```python
df = pd.read_excel('/Users/jasonchen/Documents/NYCAddresses8.xlsx')
```

```python
df['House No'] = df['House No'].astype(str)
df['Address'] = df['House No'] + ' ' + df['Street Name'] + ', ' + df['City'] + ', ' + df['State'] + ', ' + df['Country']
```

```python
from geopy.extra.rate_limiter import RateLimiter
```

```python
geocode = RateLimiter(locator.geocode, min_delay_seconds=1)
```

```python
df['Location'] = df['Address'].apply(geocode)
```

```python
df['point'] = df['Location'].apply(lambda loc: tuple(loc.point) if loc else None)
```

```python
df[['latitude', 'longitude', 'altitude']] = pd.DataFrame(df['point'].tolist(), index=df.index)
```

```python
map1 = folium.Map(
    location=[40.712776,-74.005974],
    tiles='cartodbpositron',
    zoom_start=12,
)
df=df.dropna(subset=['longitude'])
df=df.dropna(subset=['latitude'])
df.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]]).add_to(map1), axis=1)
map1
```

```python
map1.save("5million.html")
```

```python
for i,row in df.iterrows():
    lat = df.at[i, 'latitude']
    lon = df.at[i, 'longitude']
    work = df.at[i, 'Work Type']
    cost = df.at[i, 'Estimated Job Costs']
    
    if work == 'General Construction':
        color = 'blue'
    elif work == 'Mechanical Systems':
        color = 'red'
    elif work == 'Structural':
        color = 'green'
    elif work == 'Supported Scaffold':
        color = 'orange'
    elif work == 'Construction Fence':
        color = 'purple'
    elif work == 'Sidewalk Shed':
        color = 'yellow'
    else:
        color = 'brown'
    folium.Marker(location=[lat,lon], popup=[work,cost] , icon = folium.Icon(color=color)).add_to(map1)
```

```python
map2 = folium.Map(
    location=[40.712776,-74.005974],
    tiles='OpenStreetMap',  
    zoom_start=12,
)
for i,row in df.iterrows():
    lat = df.at[i, 'latitude']
    lon = df.at[i, 'longitude']
    work = df.at[i, 'Work Type']
    cost = df.at[i, 'Estimated Job Costs']
    
    if work == 'General Construction':
        color = 'blue'
    elif work == 'Mechanical Systems':
        color = 'red'
    elif work == 'Structural':
        color = 'green'
    elif work == 'Supported Scaffold':
        color = 'orange'
    elif work == 'Construction Fence':
        color = 'purple'
    elif work == 'Sidewalk Shed':
        color = 'yellow'
    else:
        color = 'black'
    
    popup = work, '${:,}'.format(cost)
    folium.Marker(location=[lat,lon], popup= popup, icon = folium.Icon(color=color)).add_to(map2)
```
